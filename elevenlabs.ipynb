```python
chapter = '6_Rationalization'
```

```python
with open(f'{chapter}/{chapter}.txt',encoding='utf-8') as f:
    text = f.read()
print(text)
```

```python
import openai
def jsonify(text):
  # Generate a response from the ChatGPT model
  inst = '''Turn the following text into a json list, with keys being speaker and content. 
  the speaker key that can take the values Eva, Alec or Narrator''' 
  completion = openai.ChatCompletion.create(
      model='gpt-3.5-turbo-16k',
        messages= [
          {'role': 'system', 'content': inst },
          {'role': 'user', 'content': text }]
  )
  return completion.choices[0].message.content
jason = jsonify(text)
```

```python
with open(f'{chapter}/{chapter}.json','w',encoding='utf-8') as f:
    f.write(jason)
```

```python
voices = {
        'Alec':'Adam',
        'Eva':'Rachel',
        'Narrator': 'Antoni',
        'Dr. Henriksson': 'Antoni',
        }
```

```python
import json
# Load and parse the JSON data from the file
def load_json(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data

# Display the parsed data
def display_data(data):
    for entry in data:
        speaker = entry["speaker"]
        content = entry["content"]
        
        # Print the speaker and content
        if speaker == "narrator":
            print(content)
        else:
            print(f"{speaker}: {content}")

filename = chapter+'/'+chapter+'.json'
parsed_data = load_json(filename)
display_data(parsed_data)
```

```python
parsed_data
```

```python
# Check that the voices are matching 
for entry in parsed_data:
    speaker = entry["speaker"]
    print(speaker)
    content = entry["content"]
    voice = voices.get(speaker, voices["Narrator"]) 
    print(voice)
    print('---')
```

```python
from elevenlabs import generate, play, save

# def generate_audio_from_json(parsed_data, output_filename):
# TODO: add output_filename optional argument
all_audios = []
n = 0

for entry in parsed_data:
    speaker = entry["speaker"]
    content = entry["content"]
    voice = voices.get(speaker, voices["Narrator"])  # Default to 'Narrator' if voice not found

    # Generate audio for the content
    audio = generate(
        text=content,
        voice=voice,
        model="eleven_monolingual_v1"
    )
    all_audios.append(audio)
    save(audio,f'{chapter}/audio/{n}_{speaker}.mp3')
    n += 1
```

```python
# Concatenate all audio segments (assuming the library returns audio in a format that can be concatenated)
final_audio = b"".join(all_audios)
    
# Save the concatenated audio to the output file
save(final_audio,f'{chapter}/audio.mp3')
# save(final_audio, output_filename)
```
